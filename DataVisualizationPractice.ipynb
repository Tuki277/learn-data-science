```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cites = pd.read_csv('california_cities.csv')
cites.head()
```

```python
#extract latd and longd

lat, long = cites['latd'], cites['longd']
population, area = cites['population_total'], cites['area_total_km2']
lat[:10]

#plot
plt.style.use('seaborn-v0_8-whitegrid')
plt.scatter(long, lat, c=np.log10(population), cmap='viridis', s=area, linewidths=0, alpha=0.4)
plt.axis('equal')
plt.colorbar(label="S")
plt.ylabel("latd")
plt.xlabel("longd")
plt.title("california cities")

##  create a legends for cities
area_range = [50, 100, 300, 500]

for area in area_range:
    plt.scatter([], [], s=area, label=str(area) + 'km$^2$', c='k')

plt.legend(scatterpoints=1, labelspacing=1, title="city arena")
```
